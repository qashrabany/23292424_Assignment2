```python
# Task Manager Program - Corrected Version with Comments and Doctests
# Techniques used: Lists, Loops, Functions, Basic Error Handling (Week 6)

# Global list to store tasks
tasks = []

def add_task(task):
    """
    Adds a task to the tasks list.

    >>> tasks.clear()  # Reset list before test
    >>> add_task("Read book")
    Added: Read book
    >>> tasks[-1]
    'Read book'
    """
    # Append the new task to the global list
    tasks.append(task)
    print(f"Added: {task}")  # Inform the user

def remove_task(index):
    """
    Removes a task by zero-based index.

    >>> tasks.clear()
    >>> add_task("Test task")
    Added: Test task
    >>> remove_task(0)
    Removed: Test task
    >>> remove_task(10)
    Invalid index!
    """
    # Check if index is within valid range
    if 0 <= index < len(tasks):
        removed = tasks[index]  # Store task for message
        del tasks[index]        # Remove task
        print(f"Removed: {removed}")
    else:
        # Handle invalid index gracefully
        print("Invalid index!")

def list_tasks():
    """
    Lists all tasks in the tasks list.

    >>> tasks.clear()
    >>> add_task("Task 1")
    Added: Task 1
    >>> list_tasks()
    1: Task 1
    """
    if not tasks:
        # No tasks available
        print("No tasks available.")
    else:
        # Enumerate tasks to show index (1-based) and description
        for i, t in enumerate(tasks):
            print(f"{i+1}: {t}")

def main():
    """Main program demonstrating task manager functionality."""

    # Add initial tasks
    add_task("Buy milk")
    add_task("Pay bills")
    add_task("Walk dog")

    # Display current tasks
    print("\nCurrent Tasks:")
    list_tasks()

    # Remove second task (index 1, zero-based indexing)
    print("\nRemoving task at index 1:")
    remove_task(1)

    # Display tasks after removal
    print("\nTasks after removal:")
    list_tasks()

# Run doctests and main program when script is executed
if __name__ == "__main__":
    import doctest
    doctest.testmod()  # Run doctests to validate functions
    main()             # Execute main task manager workflow
```

```python
import requests

def refined_safe_weather_data_fetch(city):
    """
    Fetch weather data with basic error handling - Week 8 style

    >>> refined_safe_weather_data_fetch("")
    Error: City name cannot be empty
    >>> refined_safe_weather_data_fetch("InvalidCity123")  # doctest: +SKIP
    Error: Could not connect to weather service
    >>> data = refined_safe_weather_data_fetch("London")
    >>> isinstance(data, dict)
    True
    """
    if not city:
        print("Error: City name cannot be empty")
        return None

    try:
        url = f"http://wttr.in/{city}?format=j1"
        response = requests.get(url)
        data = response.json()

        try:
            current = data['current_condition'][0]
            weather_info = {
                'city': city,
                'temperature': current['temp_C'],
                'wind_speed': current['windspeedKmph'],
                'description': current['weatherDesc'][0]['value']
            }
            return weather_info
        except:
            print("Error: Could not extract weather data from response")
            return None
    except:
        print("Error: Could not connect to weather service")
        return None

# Optional doctest check
if __name__ == "__main__":
    import doctest
    doctest.testmod()

# **Manual calls to see output in Colab**
print(refined_safe_weather_data_fetch("London"))
refined_safe_weather_data_fetch("")
```
